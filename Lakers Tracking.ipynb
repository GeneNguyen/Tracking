# DRIVES
***

```python
# Load data
drives = pd.read_excel('Tracking.xlsx', sheet_name='Drives')
```

```python
# Plot
ax = sns.lineplot(x="GAME", y="DRIVES", hue="PLAYER", legend = 'full', data=drives)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
ax.set_title('Drives')

# Legend
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.85, box.height]) # resize position
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)

# Size
fig = plt.gcf()
fig.set_size_inches(14, 7)
```

```python
# Facet grid
sns.set(style="ticks", color_codes=False)
g = sns.FacetGrid(drives, col="PLAYER", col_wrap=5, height=3, aspect = 4, sharex = True)
g = g.map(plt.plot, "GAME", "DRIVES", marker=".")

for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.xaxis.set_major_formatter(ticker.ScalarFormatter())

fig = plt.gcf()
fig.set_size_inches(12, 8)
```
